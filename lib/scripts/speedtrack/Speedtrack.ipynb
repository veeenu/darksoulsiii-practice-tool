```python
import pandas as pd
import numpy as np

df = pd.read_csv('data.txt', sep=',')
df['distance'] = np.sqrt(df['x'].diff(1).pow(2.) + df['y'].diff(1).pow(2) + df['z'].diff(1).pow(2)) / df['igt'].diff(1)
df = df.set_index(pd.to_timedelta(df['igt'] * 1000000))
df
```

```python
ax = df['distance'].plot(figsize=(16, 9))
ax.figure.savefig('fig.png')
```
